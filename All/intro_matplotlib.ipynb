## Matplotlib example workflow

# Types
* ax.plot(x, y)
* ax.scatter(x, y)
* ax.bar(x, height)
* ax.barh(y, width)
* ax.hist(x)
* fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2)

```python
# Inporting
import numpy as np, matplotlib.pyplot as plt
from math import sin
# Prepare data
x = np.arange(2000, step = 70)

# Setup lot
fig, ax = plt.subplots(figsize = (4, 4)) # width, height

# Plot data
ax.plot(x, np.sin(x))
# Customizing
ax.set(title = "Sinus function",
       xlabel = "x",
       ylabel = "sin(x)")

# Save

fig.savefig("plots/sinus.png")
```
